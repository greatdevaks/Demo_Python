```python
%matplotlib notebook

import matplotlib.pyplot as plt

x1_values = [1, 2, 3, 4]
y1_values = [10, 20, 30, 40]

x2_values = [4, 5, 2, 8]
y2_values = [20, 11, 20, 40]

x3_values = [11, 23, 83, 114]
y3_values = [101, 220, 319, 140]

plt.plot(x1_values, y1_values, "bs")
plt.plot(x2_values, y2_values, "ro")
plt.plot(x3_values, y3_values, "g--")
plt.xlabel("My X Label: Number of Cars")
plt.ylabel("My Y Label: Prices")
plt.show()
```
